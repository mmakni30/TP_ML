## 1. Logistic regression, IRLS algorithm.

### Preliminary question: the algorithm

#### Have a look at the function `regression_logistique.m` and locate the main steps of the algorithm you have been taught (see course).
You can comment the code in the cell below.

```python
def regression_logistique(X,t,Nitermax=20,eps_conv=1e-3):
    '''Entrees :
    X = [ones(N_train,1) x_train];
    t = class_train 
    Nitermax = nombre maximale d'itérations (20 par défaut)
    eps_conv = critère de convergence sur norm(w-w_old)/norm(w) ; 
    eps_conv=1e-3 par défaut
    
    Sorties : 
    w : vecteur des coefficients de régression logistique
   Niter : nombre d'itérations utilisées effectivement
   
   Fonction de régression logistique pour la classification binaire.
   
   Utilisation :
       Nitermax = 50
       eps_conv = 1e-4
       [w,Niter] = regression_logistique(X,t,Nitermax,eps_conv)
    '''
    N_train = X.shape[0]

    #initialisation : 1 pas de l'algorithme IRLS
    w = np.zeros((X.shape[1],))
    w_old = w 
    y = 1/2*np.ones((N_train,))
    R = np.diag(y*(1-y))   # diag(y_n(1-y_n))
    z = X.dot(w_old)-la.inv(R).dot(y-t)
    w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)

    # boucle appliquant l'algortihme de Newton-Raphson
    Niter = 1
    while ( (la.norm(w-w_old)/la.norm(w)>eps_conv) | (Niter<Nitermax) ):
        Niter = Niter+1
        y = 1/(1+np.exp(-X.dot(w))) # Apply sigmoid function to XwT
        R = np.diag(y*(1-y)) # construction of the diagonal matrix y(1-y)
        w_old = w 
        z = X.dot(w_old)-la.inv(R).dot(y-t) # calculate z using old values of w 
        w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z) # calcul new w
         
    return w, Niter
```

### Exercise 1 : the logistic function of decision

1. Use the function `regression_logistique.m` to estimate the logistic
regression vector `w`. *Indication : use `Nitermax = 50;
eps_conv=1e-3;`.*
2. Compute the decision function $f(x) = argmax_k P(C_k|x)$ on the test set
to get the classification results. Recall that $y_n=\sigma(w^T x)$ (logistic function)
and that *using vectors* you may directly write $y=\sigma(Xw)$, with the
column of ones in X.
3. Display the results by plotting the points from both the training set
and the test set.
4. Write the equation which defines the decision boundary.
5. Artificially add a few points to the training set far from the decision boundary to check the robustness of logistic regression to outliers. Check the behaviour of LDA for comparison in this case and comment. 


```python
#Exercise 1
w, Niter = regression_logistique(X, t, Nitermax=50, eps_conv=1e-3)
print("final coefficients: ", w)
sigmoid = lambda x : 1/(1 + np.exp(-x))
decision = lambda x, w: argmax([1 - sigmoid(np.dot(x, w.T)), sigmoid(np.dot(x, w.T))], axis=0)
def error(pred, truth):
    """Calculate the average error"""
    return (np.sum([pred != truth]))/truth.shape[0]
```

You can draw the decision boundary $w^Tx = 0$ by using: 


```python
pred = decision(X_test, w)
pred = np.array(pred)
plt.figure(figsize=(6, 4), facecolor='white') 
x1 = np.linspace(-2.5,1.5,10) 
x2 = (-w[0]-w[1]*x1)/w[2]
plt.plot(x1,x2,'k--')
# plot the right predictions and the misclassified
plt.scatter(X_test[np.where((pred==0) & (t_test==0)),1], X_test[np.where((pred==0) & (t_test==0)),2], color="b", label="class 1")
plt.scatter(X_test[np.where((pred==1) & (t_test==1)),1], X_test[np.where((pred==1) & (t_test==1)),2], color="r", label="class 2")
plt.scatter(X_test[np.where(pred - t_test !=0),1], X_test[np.where(pred - t_test !=0),2], color="black", label="Misclassified")

plt.xlabel("First attribute")
plt.ylabel("Second attribute")
plt.legend()
plt.title("Classes predicted from training set")
plt.show()
```

As said previously, we have few misclassifications given the points near the decision line.

```python
pred_test = decision(X_test, w)
cm = confusion_matrix(pred_test, t_test)
tn, fp, fn, tp = cm.ravel()
plt.figure(figsize=(8, 5), facecolor='white')
plt.title("LR Confusion Matrix")
sns.heatmap(cm, annot=True, cbar=False)
plt.xlabel("Truth")
plt.ylabel("Predicted")
plt.show()
```

We have a difficulty predicting the minority class (class 1) and this is a common problem for logistic regression models, we might want to augmentate the data (oversample it) and undersample it. The logistic regression models are usually more robust on larger data set, so undersampling the set will decrease the performance.

```python
precision_lr = tp / (tp + fp)
recall_lr = tp / (tp + fn)
F1_lr = 2 * (precision_lr * recall_lr) / (precision_lr + recall_lr)
mcc_lr = (tp*tn - fp*fn)/ sqrt((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn))
print(f"The precision of LR = {round(precision_lr, 4)}")
print(f"The recall of LR = {round(recall_lr, 4)}")
print(f"The F1 of LR = {round(F1_lr, 4)}")
print(f"The MCC score of LR = {round(mcc_lr, 4)}")
```

Here, we have a good recall so the classifier returns most of the relevant results: it means that it classify well the two classes.

a less good precision means that the classifier returns a lot of false positives which is equal to 10 and it's due to the umbalanced data.

On the decision boundary, the output of $\sigma(w^Tx_n)=0.5$

So $w^Tx_n=0$

$w^T$ $x$ = 0 $\Leftrightarrow$ $w_{0} + w_{1}x_{1} + w_{2}x_{2}$ = 0  $\Leftrightarrow$ $x_{2}$ = $-\frac{w_{0}}{w_{2}}$ $-\frac{w_{1}}{w_{2}}$  $x_{1}$

```python
a = -w[1]/w[2]
b = -w[0]/w[2]

print(f"The decision boundary equation : y = {round(a,2)} x + {round(b,2)}")
```

### Logistic regression with outliers
Here we will apply our model on a data that contains outliers. They should be easier to classify as they are fare from the decision boundary, but this would mean that in our diagonal matrix containing the y(1 -y) there will be zeros, so the matrix might not be inversible

```python
def regression_logistique_not_inv_R(X,t,Nitermax=20,eps_conv=1e-3, out=False):
    '''Entrees :
    X = [ones(N_train,1) x_train];
    t = class_train 
    Nitermax = nombre maximale d'itérations (20 par défaut)
    eps_conv = critère de convergence sur norm(w-w_old)/norm(w) ; 
    eps_conv=1e-3 par défaut
    
    Sorties : 
    w : vecteur des coefficients de régression logistique
   Niter : nombre d'itérations utilisées effectivement
   
   Fonction de régression logistique pour la classification binaire.
   
   Utilisation :
       Nitermax = 50
       eps_conv = 1e-4
       [w,Niter] = regression_logistique(X,t,Nitermax,eps_conv)
    '''
    N_train = X.shape[0]

    #initialisation : 1 pas de l'algorithme IRLS
    w = np.zeros((X.shape[1],))
    w_old = w 
    y = 1/2*np.ones((N_train,))
    R = np.diag(y*(1-y))  # diag(y_n(1-y_n))
    z = X.dot(w_old)-la.inv(R).dot(y-t)
    
    w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)

    # boucle appliquant l'algortihme de Newton-Raphson
    Niter = 1
    while ( (la.norm(w-w_old)/la.norm(w)>eps_conv) | (Niter<Nitermax) ):
        Niter = Niter+1
        y = 1/(1+np.exp(-X.dot(w)))
        R = np.diag(y*(1-y))  
        w_old = w        
        try: # if the R is not inversible, we stop
            inv_R = la.inv(R)
        except:
            if out:
                print(f" R is not inversible after {Niter}")

            break
        z = X.dot(w_old)-inv_R .dot(y-t)
        w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)
         
    return w, Niter
```

```python
np.random.seed(10)
N_add_1, N_add_2 = 100,100
sigma_1 = (0.1, 0.5)
sigma_2 = (0.5, 0.1)

x_far_1 = np.array([-2,4]) + np.random.normal(scale=sigma_1, size=(N_add_1,2))
X_far_1 = np.hstack((np.ones((N_add_1,1)), x_far_1))
t_far_1 = np.ones((1,N_add_1))[0]

x_far_2 = np.array([1,0]) + np.random.normal(scale=sigma_2, size=(N_add_2,2))
X_far_2 = np.hstack((np.ones((N_add_2,1)),x_far_2))
t_far_2 = np.zeros((1,N_add_2))[0]

x_far = np.concatenate([x_far_1,x_far_2])
X_far = np.concatenate([X_far_1,X_far_2])
t_far = np.concatenate([t_far_1,t_far_2])

x_train_new = np.concatenate([x_train,x_far])
X_new = np.concatenate([X,X_far])

class_train_new = np.concatenate([t,t_far])

t_new = 2 - class_train_new
w_out, Niter = regression_logistique_not_inv_R(X_new, t_new, out=True)
pred_new = decision(X_new, w_out)
print("final coefficients: ", w)
```

```python
print("error on the training set with outliers: ", round(error(pred_new, t_new), 4))
print("error on the test set with outliers: ", error(decision(X_test, w_out), t_test))
```

The matrix was not inversible after 3 iterations, so we added a breaking point in our function to stop the iterations when the R matrix is no longer inversible. This means that given these outliers, some elements in the R matrix will converge to zero really fast, even if the algorithm hasn't reached a good estimation of the intercept. The model will then fail to classify well our observations.

```python
pred_out = decision(X_new, w_out)
plt.figure(figsize=(15, 10), facecolor='white') 
# plot the prediction on the new data with outliers
plt.subplot(2, 2, 1)
plt.scatter(X_new[np.where((pred_out==0) & (t_new==0)),1], X_new[np.where((pred_out==0) & (t_new==0)),2], color="b", label="class 1")
plt.scatter(X_new[np.where((pred_out==1) & (t_new==1)),1], X_new[np.where((pred_out==1) & (t_new==1)),2], color="r", label="class 2")
plt.scatter(X_new[np.where(pred_out - t_new !=0),1], X_new[np.where(pred_out - t_new !=0),2], color="black", label="Misclassified")
plt.xlabel("First attribute")
plt.ylabel("Second attribute")
plt.legend()
plt.title("Classes predicted from training set")

# plot the new data with outliers
plt.subplot(2, 2, 2)
x1 = np.linspace(-2.5,2.5,10) 
x2 = (-w[0]-w[1]*x1)/w[2]
class_train_1_n = np.where(class_train_new==0)[0]
class_train_2_n = np.where(class_train_new==1)[0]
plt.scatter(x_train_new[class_train_1_n][:,0], x_train_new[class_train_1_n][:,1], color="b", label="class 1")
plt.scatter(x_train_new[class_train_2_n][:,0], x_train_new[class_train_2_n][:,1], color="r", label="class 2")
plt.xlabel("First attribute")
plt.ylabel("Second attribute")
plt.legend()
plt.title("Classes plot from training set with outliers")
plt.show()
```

Since the estimation of the coefficients is wrong, we didn't plot the decision boundary but chose to plot the misclassified observations. Now class 1 was well classified. From the coefficients array we can notice that there is a big gap between the weight correponding to feature 1 and the intercept and that of the second features (which is a negative value). 

Class 2 contains mostly element whose first feature coordinate is negative, which means $-w_1x_{k,1}$ in the probability expression for class 1 will make it so that $P(x_k \in C1)$ converges to 0, but we should have the opposite effect with elements of class 1 (positive first coordinate) which is not the case and that might be due to the intercept that is overset. 

Logistic Regression models are not much impacted due to the presence of outliers because the sigmoid function tapers the outliers. But the presence of extreme outliers may somehow affect the performance of the model and lowering the performance.
In general, when we have extreme outliers and we want to use Logistic regression, we have to remove these points. The idea of removing the extreme misclassified data points are being implemented in the following steps: 

1. Train a logistic regression model for a given dataset
2. Compute the weight vector for the model trained in step 1. In scikit-learn, the weight vector can be computed using classifier.coef_.
3. Compute the distance of each of the data points from the weight vector.
4. Decide a distance threshold, to remove the extreme data points.
5. Remove the data points having a distance greater than the decided threshold.
6. Re-train the same logistic regression model on the new data from step 5.

### Adjusting the classification threshold

We have two times the number of obersavtions belonging to class 2 than to class 1, and this affect the calculations of the intercept in our Logistic model. One way to counter this problem, is to apply a weighted class intercept, as implemented in scikit-learn, and we will explore this part afterwards. Another way is to adjust the classification threshold, up to now we took the maximum of probabilities. That is to say we compared probabilities to 0.5. However, and especially in imbalanced cases, a different threshold might work better, by helping the classification of the minority class. 

In this part, we will take a look at the probability distribution and plot the ROC curve to get the best threshold (maximizing the G-mean score, $g-mean = \sqrt{TPR*(1-FPR)}$)

```python
unique, counts = np.unique(class_test, return_counts=True)
print(f"The balance of the test data {dict(zip(unique, counts))}")
```

```python
from sklearn.metrics import roc_curve

w, _ = regression_logistique(X, t)
pred_proba = sigmoid(np.dot(X, w.T))
fpr, tpr, thresholds = roc_curve(t, pred_proba)
best_thres = np.argmax(np.sqrt(tpr * (1 - fpr)))
plt.figure(figsize=(15, 9), facecolor="white")
plt.subplot(2, 2, 1)
plt.plot(fpr, tpr)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC curve on the training set")
plt.scatter(fpr[best_thres], tpr[best_thres], label=f"best threshold:{thresholds[round(best_thres, 4)]}")
plt.legend()
plt.subplot(2, 2, 2)
plt.hist(pred_proba, bins = 50)
plt.title("histogram of probabilities on training set")
plt.show()
```

Most probabilities are distributed around 0 given the class imbalance, so it makes sense that the best threshold we be lower than 0.5, and in this case 0.27. Usually theROC curve could be optimistic and doesn't take into account the performance on the minority class. We will use the F1-MCC curve that is more suitable for imbalanced data such as ours [4], the best threshold will be the closest to the point (1, 1):

```python
from sklearn.metrics import f1_score, matthews_corrcoef, roc_curve
# function to compare the probabilities with a threshold given as an argument
decision_thres = lambda x, thres: np.array([thres<=sigmoid(np.dot(x, w.T))]).astype(int)

F1 = np.zeros(len(thresholds))
MCC = np.zeros(len(thresholds))
# in order not to store a matrix of prediction for each threshold, 
# we'll iterate over the thresholds
for idx, thres in enumerate(thresholds):
    y_pred = decision_thres(X, thres).reshape(t.shape)
    F1[idx] = f1_score(t, y_pred)
    MCC[idx] = matthews_corrcoef(t, y_pred)
best = np.array([1, 1])
best_thres = np.argmin([la.norm(best -x) for x in np.hstack((F1, MCC))])
plt.figure(figsize=(15, 9), facecolor="white")
plt.subplot(2, 2, 1)
plt.plot(F1, MCC)
plt.xlabel("F1")
plt.ylabel("MCC")
plt.title("F1-MCC curve")
plt.scatter(MCC[best_thres], F1[best_thres], label=f"best threshold:{thresholds[round(best_thres, 4)]}")
plt.legend()
plt.show()
```

```python
y_pred = decision_thres(X_test, 0.273).reshape(t_test.shape)
cm = confusion_matrix(y_pred, t_test)
tn, fp, fn, tp = cm.ravel()
plt.figure(figsize=(8, 5), facecolor='white')
plt.title("LR Confusion Matrix with optimized threshold")

sns.heatmap(cm, annot=True, cbar=False)
plt.xlabel("Truth")
plt.ylabel("Predicted")
plt.show()
```

The confusion matrix plots the decision on the test set. We have better classification on the class 1 than before (here class 1), as opposed to the classification of class 2 (0), which makes sense because we allowed more instances o be considered True. In our case, F1-MCC curve and the ROC curve have given similar best thresholds, we won't need to weight their outputs in order to find the best one.

### Final scores

```python
precision_lr = tp / (tp + fp)
recall_lr = tp / (tp + fn)
F1_lr = 2 * (precision_lr * recall_lr) / (precision_lr + recall_lr)
mcc_lr = (tp*tn - fp*fn)/ sqrt((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn))
print(f"The precision of LR = {round(precision_lr, 4)}")
print(f"The recall of LR = {round(recall_lr, 4)}")
print(f"The F1 of LR = {round(F1_lr, 4)}")
print(f"The MCC score of LR = {round(mcc_lr, 4)}")
```

We can conclude that the Logistic regression works really well in this situation and we won't need to go for more complex non-linear models.

### Comparing the results to LDA model
In this part we will compare our Logistic model to an LDA model, with wand without the outliers addition. As we know, the LDA takes into account the whole distribution of the data so it might be affect by the presence of points far from the normal distribution it expect it to be.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

LinearClassifier = LinearDiscriminantAnalysis(store_covariance=True)
LinearClassifier.fit(X=x_train,y=t)

LinearClassifier_new = LinearDiscriminantAnalysis(store_covariance=True)
LinearClassifier_new.fit(X=x_train_new, y=t_new)


t_test_pred_without_outliers = LinearClassifier.predict(x_test)
test_error_without_outliers = error(t_test_pred_without_outliers,t_test)

t_test_pred_with_outliers = LinearClassifier_new.predict(x_test)
test_error_with_outliers = error(t_test_pred_with_outliers,t_test)

print(f"Test error without outliers = {test_error_without_outliers}")
print(f"Test error with outliers = {test_error_with_outliers}")
```

#### Function to plot LDA
Extracted from the sk-learn documentation [5]

```python
cmap = colors.LinearSegmentedColormap(
    "red_blue_classes",
    {
        "red": [(0, 1, 1), (1, 0.7, 0.7)],
        "green": [(0, 0.7, 0.7), (1, 0.7, 0.7)],
        "blue": [(0, 0.7, 0.7), (1, 1, 1)],
    },
)
plt.cm.register_cmap(cmap=cmap)
```

```python
def plot_data(lda, X, y, y_pred, fig_index):
    splot = plt.subplot(2, 2, fig_index)
    if fig_index == 1:
        plt.title("Linear Discriminant Analysis")
        plt.ylabel("Data with\n fixed covariance")
    elif fig_index == 3:
        plt.ylabel("Data with\n varying covariances")

    tp = y == y_pred  # True Positive
    tp0, tp1 = tp[y == 0], tp[y == 1]
    X0, X1 = X[y == 0], X[y == 1]
    X0_tp, X0_fp = X0[tp0], X0[~tp0]
    X1_tp, X1_fp = X1[tp1], X1[~tp1]

    # class 0: dots
    plt.scatter(X0_tp[:, 0], X0_tp[:, 1], marker=".", color="red")
    plt.scatter(X0_fp[:, 0], X0_fp[:, 1], marker="x", s=20, color="#990000")  # dark red

    # class 1: dots
    plt.scatter(X1_tp[:, 0], X1_tp[:, 1], marker=".", color="blue")
    plt.scatter(
        X1_fp[:, 0], X1_fp[:, 1], marker="x", s=20, color="#000099"
    )  # dark blue

    # class 0 and 1 : areas
    nx, ny = 200, 100
    x_min, x_max = plt.xlim()
    y_min, y_max = plt.ylim()
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx), np.linspace(y_min, y_max, ny))
    Z = lda.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    Z = Z[:, 1].reshape(xx.shape)
    plt.pcolormesh(
        xx, yy, Z, cmap="red_blue_classes", norm=colors.Normalize(0.0, 1.0), zorder=0, shading='auto' 
    )
    plt.contour(xx, yy, Z, [0.5], linewidths=2.0, colors="white")

    # means
    plt.plot(
        lda.means_[0][0],
        lda.means_[0][1],
        "*",
        color="yellow",
        markersize=15,
        markeredgecolor="grey",
    )
    plt.plot(
        lda.means_[1][0],
        lda.means_[1][1],
        "*",
        color="yellow",
        markersize=15,
        markeredgecolor="grey",
    )

    return splot


def plot_ellipse(splot, mean, cov, color):
    v, w = linalg.eigh(cov)
    u = w[0] / linalg.norm(w[0])
    angle = np.arctan(u[1] / u[0])
    angle = 180 * angle / np.pi  # convert to degrees
    # filled Gaussian at 2 standard deviation
    ell = mpl.patches.Ellipse(
        mean,
        2 * v[0] ** 0.5,
        2 * v[1] ** 0.5,
        180 + angle,
        facecolor=color,
        edgecolor="black",
        linewidth=2,
    )
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(0.2)
    splot.add_artist(ell)
    splot.set_xticks(())
    splot.set_yticks(())


def plot_lda_cov(lda, splot):
    plot_ellipse(splot, lda.means_[0], lda.covariance_, "red")
    plot_ellipse(splot, lda.means_[1], lda.covariance_, "blue")
```

```python
i = 0
plt.figure(figsize=(10, 8), facecolor="white")
plt.suptitle(
    "Linear Discriminant Analysis with and without outliers",
    y=0.98,
    fontsize=15,
)

# Linear Discriminant Analysis without ourliers
splot = plot_data(LinearClassifier, x_test, t_test, t_test_pred_without_outliers, fig_index=1)
plot_lda_cov(LinearClassifier, splot)
plt.axis("tight")
# Linear Discriminant Analysis with ourliers

splot = plot_data(LinearClassifier_new, x_test, t_test, t_test_pred_with_outliers, fig_index=2)
plot_lda_cov(LinearClassifier_new, splot)
plt.axis("tight")
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```

In these plots, we can clearly see the influence of outliers on LDA. The covariance matrix is completely different(eigenvalues increase and eigenvectors are changed). The eclipses are too far from the decision boundary due to the far outliers.

We can also see that points from class 1 are not well classified which implies that outliers of class 1 are far from the decision boundary.

We have to deal with outliers before using LDA in order to have a good classifier.

```python
cm = confusion_matrix(t_test_pred_without_outliers, t_test)
tn, fp, fn, tp = cm.ravel()
precision_lr = tp / (tp + fp)
recall_lr = tp / (tp + fn)
F1_lr = 2 * (precision_lr * recall_lr) / (precision_lr + recall_lr)
mcc_lr = (tp*tn - fp*fn)/ sqrt((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn))
print(f"The precision of LDA = {round(precision_lr, 4)}")
print(f"The recall of LDA = {round(recall_lr, 4)}")
print(f"The F1 of LDA = {round(F1_lr, 4)}")
print(f"The MCC score of LDA = {round(mcc_lr, 4)}")
```

As expected, the LDA failed when we add the outliers as the error on the test set goes up to 77.5%, but it performed better than our best logistic regression model on the data with and without the outliers. This could stem from the fact that we added extreme outliers which stopped the learning process of the LR (R is no longer inversible). Also, as we have seen in the last work (on LDA and QDA), this data was linearly separable and normally distributed, in this case using a LDA model would be more pertinent, even though both models didn't have good results with the outliers addition. 

Just like the previous work, we could change the representation of the data to $\phi(X)$ so that it is linearly separable.

We can use more robust algorithm that can handle outliers such as RLDA, one-step M-estimator (MOM) and winsorized one-step M-estimator (WMOM) to alleviate the classification problem. These two robust estimators(MOM and WMOM) used trimming and winsorizing approach to eliminate the outliers of the data and then form the robust discriminant rule. Their function as the substitutes for the classical estimators in Linear Discriminant Analysis (LDA) model very much improves the misclassification rates.[1]

```python
list_sig = np.where(std > 0.7)[0]
x_train = x_train_brut[:, list_sig]
x_test = x_test_brut[:, list_sig]
X_train = np.hstack((np.ones((N_train,1)),x_train))
X_test = np.hstack((np.ones((x_test.shape[0],1)),x_test))
w, Niter = regression_logistique_not_inv_R(X_train, class_train, 20, 1e-3, out=True)
y_pred_test = decision(X_test, w)
print("new error on test set: ", round(error(y_pred_test, class_test), 4))
```

We have the best model with a higher threshold than the one set at the beginning. At each iteration we have a different value, we can either set a random state, or apply a cross validation to get a more robust result. 

The training error is 0 for all thresholds, which suggests that we are overfitting. The next step would be to add a regularisation factor, with the l1 norm, we will apply a grid search on this parameter as well with values from 1e-8 to 100:

```python
def regression_logistique_regularized(X,t,Nitermax=20,eps_conv=1e-3, Lambda = 1e-5):
    '''Entrees :
    X = [ones(N_train,1) x_train];
    t = class_train 
    Nitermax = nombre maximale d'itérations (20 par défaut)
    eps_conv = critère de convergence sur norm(w-w_old)/norm(w) ; 
    eps_conv=1e-3 par défaut
    Lambda = 1e-5 facteur de régularisation
    
    Sorties : 
    w : vecteur des coefficients de régression logistique
   Niter : nombre d'itérations utilisées effectivement
   
   Fonction de régression logistique pour la classification binaire.
   
   Utilisation :
       Nitermax = 50
       eps_conv = 1e-4
       Lambda = 1e-5
       [w,Niter] = regression_logistique(X,t,Nitermax,eps_conv)
    '''
    N_train = X.shape[0]

    #initialisation : 1 pas de l'algorithme IRLS
    w = np.zeros((X.shape[1],))
    w_old = w 
    y = 1/2*np.ones((N_train,))
    R = np.diag(y*(1-y)) + Lambda*np.identity(len(y)) # diag(y_n(1-y_n))
    z = X.dot(w_old)-la.inv(R).dot(y-t)
    
    w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)

    # boucle appliquant l'algortihme de Newton-Raphson
    Niter = 1
    while ( (la.norm(w-w_old)/la.norm(w)>eps_conv) | (Niter<Nitermax) ):
        Niter = Niter+1
        y = 1/(1+np.exp(-X.dot(w)))
        R = np.diag(y*(1-y))  
        w_old = w 
        inv_R = la.inv(R+2*Lambda*np.eye(R.shape[0]))
        z = X.dot(w_old)-inv_R .dot(y-t)
        w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)
         
    return w, Niter
```

```python
X_train_t, X_val, class_train_t, class_val = train_test_split(X_train, class_train, test_size=0.2)
pred_lambda = lambda l: error(decision(X_train_t, regression_logistique_regularized(X_train_t, class_train_t, 20, 1e-3, Lambda=l)[0]), class_train_t)
pred_lambda_val = lambda l: error(decision(X_val, regression_logistique_regularized(X_train_t, class_train_t, 20, 1e-3, Lambda=l)[0]), class_val)

lambda_list = np.logspace(-8, 2, 10)
lambda_train_error = list(map(pred_lambda, lambda_list))
lambda_val_error = list(map(pred_lambda_val, lambda_list))
```

```python
plt.figure(figsize=(7, 4), facecolor="white")
plt.plot(lambda_list, lambda_train_error, color="b", label="training")
plt.plot(lambda_list, lambda_val_error, color="r", label="test")
plt.scatter(lambda_list[np.argmin(lambda_val_error)], lambda_val_error[np.argmin(lambda_val_error)], label=f"Lambda={lambda_list[np.argmin(lambda_val_error)]}")
plt.legend()
plt.title("Error on the training set from LR with regularisation")
plt.show()
```

```python
w_lambda, N_iter = regression_logistique_regularized(X_train, class_train, 20, 1e-3, Lambda=100)
pred_test_lambda = decision(X_test, w_lambda)
print("error with regularization on the test set: ", error(pred_test_lambda, class_test))
```

The classification results improved slightly with the l1 regularization term, the bigger the value the better the results on the validation set. We were indeed in an overfitting situation, as the test error dropped to 0.059 from 0.077 with Lambda =100. With values in the matrices ranging from -1 to 1, the regularization term shouldn't have this magnitude. Maybe the l1 penamty is not suitable for our task, in the sci-kit learn libray, the default penalty is l2 so we will see if it improves the results.

The sci-kit learn library offers multiple solvers for the optimization problem. We will test the quasi newton method lbfgs [2] and the liblinear (a linear descient along coordinate directions) [3]. The Quasi-Newton method uses the same algorithm as the New gradient descent without having to calculate the hessian matrix as it approximates it. It is the default solver in the newer version of sk-learn

Since the `lbfgs` doesn't implement the l1 penalty, we first saw which solver work best before comparing l1 and l2. We also tune the model on its regularization parameter C with a cross validation with stratified 10 folds with shuffling the data:

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
params = {"C": np.linspace(0.01, 100), "solver":["liblinear", "lbfgs"]}
lr = LogisticRegression()
grid = GridSearchCV(lr, param_grid=params ,cv=StratifiedKFold(n_splits=10, shuffle=True, random_state=10))
grid.fit(x_train, class_train)
print(f"best parameters: {grid.best_params_} and best score (error): {1 - grid.best_score_}")
```

```python
params = {"penalty": ["l1", "l2"]}
lr = LogisticRegression(C=2.051, solver="liblinear")
grid = GridSearchCV(lr, param_grid=params ,cv=StratifiedKFold(n_splits=10, shuffle=True, random_state=10))
grid.fit(x_train, class_train)
print(f"best parameters: {grid.best_params_} and best score (error): {1 - grid.best_score_}")
```

```python
best_lr = grid.best_estimator_
lr.fit(x_train, class_train)
predictions = lr.predict(x_test)
cm = confusion_matrix(predictions, class_test)
tn, fp, fn, tp = cm.ravel()
plt.figure(figsize=(8, 5), facecolor='white')
plt.title("LR (sklearn) Confusion Matrix")
sns.heatmap(cm, annot=True, cbar=False)
plt.show()
```

Commentaires :

The misclassification dropped from 25 and 44 to 15 and 18 respectively after tuning the model along the regularization parameter, the solver and the penalty. The error on the training set is still 0 which is odd, as usually the regularization decreases the performance on this set. Instead of calculating the Hessian matrix (like in our algorithm) and finding non inversible matrices if some features are colinear or some observations are far from the decision line, sci-kit learn uses another solver that approximate this matrix for the optimization part. The class activation map shows that the edges on the left play the biggest role in the classification, which makes since since that is the part the differentiate between a 5 and a 6. Most of the other pixels are not used as they don't add relevant information to this classification task.

In their documentation, liblinear works better for small datasets which is our case (around 300 samples) and the l1 penalty was better which disproves our claim earlier. However, it could be that this penalty is better for this optimization algorithm, and might not work as well for others.

This algorithm was able to classify well images in less than 0.1s and without calculation accelerators (like GPU). Usually in this task, the first model we use is a deep learning CNN model which needs a lot more samples to estimate its parameters, and it needs a lot more time to fit to the training data. Our approach should be to first look at the data available, and try simple and interpretable models like the logistic regression before going to more complicated ones.
